### Writing a compact program :-

```python
import pandas as pd
import requests

def download_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.content
    else:
        raise Exception(f"Failed to download data. Status code: {response.status_code}")

def process_data(data):
    # reading by pandas json reader as the data is in the json format
    df = pd.read_json(data)
    
    # extracting the data from the pokemon column to expand 
    data = df.pokemon
    
    # expanding the data and creating a new dataframe
    dn = pd.json_normalize(data)
    
    return dn

def convert_to_excel(dataframe, output_file):
    dataframe.to_excel(output_file, index=False)

# link to the data we want to download
data_link = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"

try:
    # Download the data
    downloaded_data = download_data(data_link)
    
    # Process the data
    processed_data = process_data(downloaded_data)
    
    # Convert to Excel and save the file
    excel_file = "output.xlsx"
    convert_to_excel(processed_data, excel_file)
    
    print(f"Data downloaded, processed, and saved as '{excel_file}' successfully.")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```
