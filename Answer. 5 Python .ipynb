Link - http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes

```python
import requests

def download_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception(f"Failed to download data. Status code: {response.status_code}")

def extract_episode_details(data):
    show_name = data['name']
    episodes = data['_embedded']['episodes']
    episode_details = []
    
    for episode in episodes:
        episode_name = episode['name']
        season_number = episode['season']
        episode_number = episode['number']
        airdate = episode['airdate']
        
        episode_details.append({
            'Show Name': show_name,
            'Episode Name': episode_name,
            'Season Number': season_number,
            'Episode Number': episode_number,
            'Airdate': airdate
        })
    
    return episode_details

# Provide the API link
api_link = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

try:
    # Download the data
    downloaded_data = download_data(api_link)
    
    # Extract episode details
    episode_details = extract_episode_details(downloaded_data)
    
    # Print the formatted episode details
    for episode in episode_details:
        print(f"Show Name: {episode['Show Name']}")
        print(f"Episode Name: {episode['Episode Name']}")
        print(f"Season Number: {episode['Season Number']}")
        print(f"Episode Number: {episode['Episode Number']}")
        print(f"Airdate: {episode['Airdate']}")
        print("-" * 20)
except Exception as e:
    print(f"An error occurred: {str(e)}")
```
