`````python
link = 'https://data.nasa.gov/resource/y77d-th95.json'
`````

`````python
import pandas as pd
`````

`````python
data = pd.read_json(link)
data.head()
`````

### ● Get all the Earth meteorites that fell before the year 2000

`````python
df = data.copy()
`````

`````python
df[['year']]
`````

`````python
df[['year']].describe()
`````

`````python
df.year.isnull().sum()
`````

`````python
df[df.year.isnull()]   # we can drop this NaN value
`````

`````python
df = df.drop([37])
`````

`````python
df.year.isnull().sum()
`````

`````python
'1880-01-01T00:00:00.000'.split('-')  # we can get the year by splitting by -
`````

`````python
# creating a new column having year values in YYYY format
df['yr'] = df.year.apply(lambda x : x.split('-')[0])  
`````

`````python
df.head()
`````

`````python
df[df.yr<'2000']
`````

These are all the Earth meteorites that fell before the year 2000.

### Get all the earth meteorites co-ordinates who fell before the year 1970

`````python
df2 = df[df.yr<'1970'][['geolocation','yr']]
df2
`````

`````python
df2.geolocation.isnull().sum()
`````

`````python
df2[df2.geolocation.isnull()]   # we need to remove these records, as they don't have location information
`````

`````python
rmv = df2[df2.geolocation.isnull()].index.to_list()
rmv
`````

`````python
df2 = df2.drop(rmv)
`````

`````python
df2 = df2.reset_index()
df2
`````

`````python
df2.describe(include='all')
`````

`````python
len([df2.geolocation[i]['coordinates'] for i in range(df2.shape[0])])
`````

These are all the earth meteorites co-ordinates who fell before the year 1970.
