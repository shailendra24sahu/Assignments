### ● Get all Pokemons that have less than 4 weaknesses

````python
dn.weaknesses
````

````python
lst =[]
for idx,val in enumerate(dn.weaknesses):
    if len(val)<4:
        lst.append(idx)
````

````python
dn.loc[lst]
````

### Get all Pokemons that do not have more than 2 evolutions

````python
dn[['next_evolution','prev_evolution']].describe(include='all')
````

````python
dn[['next_evolution','prev_evolution']].isnull().sum()
````

There are null values in these columns, we need to fill it with an empty dictionary so that we can calculate the len of that dictionary as zero.

````python
df=dn.copy()
````

````python
df['next_evolution']=df['next_evolution'].fillna('{}')
````

````python
df['prev_evolution']=df['prev_evolution'].fillna('{}')
````

````python
df[['next_evolution','prev_evolution']].head()
````

````python
lst = []
for i in range(df.shape[0]):
    if (len(df.next_evolution[i])+len(df.prev_evolution[i])) <=2:
        lst.append(i)
````

````python
lst
````

````python
df.iloc[lst]
````

These are all Pokemons that do not have more than 2 evolutions.

### ● Get all Pokemons whose spawn time is less than 300 seconds.
Note - spawn time format is "05:32”, so assume “minute: second” format and perform the analysis.

````python
dn = df.copy()
````

````python
dn[['spawn_time']].head()
````

````python
dn[['spawn_time']].describe()
````

````python
# replacing N/A values by 0:0
dn.spawn_time = dn.spawn_time.replace('N/A','0:0')  
````

````python
dn.spawn_time.isnull().sum()
````

````python
# add two new columns for time in minutes and seconds
dn['sp_min']=0
dn['sp_sec']=0
for i in range(dn.shape[0]):
    a,b = dn.spawn_time[i].split(":")
    dn['sp_min'][i]=a
    dn['sp_sec'][i]=b
````

````python
dn
````

````python
# make a new column having time in total seconds
dn['sp_time'] = dn['sp_min']*60+dn['sp_sec']
````

````python
dn[dn.sp_time<300]
````

These are all Pokemons whose spawn time is less than 300 seconds.
